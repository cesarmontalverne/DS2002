### Loading Data with an API and Pushing to MongoDB
Swagger Documentation: https://dadosabertos.camara.leg.br/swagger/api.htmls

```python
#connect to mongoDB
atlas_cluster_name = "sandbox.zibbf"
atlas_user_name = "root_user"
atlas_password = "Eutocomfome,meuamigo!"

conn_str = {"local" : f"mongodb://localhost:27017/",
    "atlas" : f"mongodb+srv://{atlas_user_name}:{atlas_password}@{atlas_cluster_name}.mongodb.net"
}

print(f"Local Connection String: {conn_str['local']}")
```

```python
# get data through api
base = "https://dadosabertos.camara.leg.br/api/v2"

#create urls
partidos_url = base+"/partidos?itens=100"
eventos_url = base+"/eventos?dataInicio=2023-01-01&dataFim=2023-03-13&ordem=ASC&ordenarPor=dataHoraInicio"
votacoes_url = base+"/votacoes"
legislaturas_url = base+"/legislaturas?itens=100&ordem=DESC&ordenarPor=id"

#make requests
def requireAPIs(url):
    try:
        req = requests.get(url)
        return req.json()['dados']
    except: 
        print(f"couldn't read {url}")
        return False


#push data to mongoDB
files = {("partidos",partidos_url), ("eventos",eventos_url),("votacoes",votacoes_url),("legislaturas",legislaturas_url)}
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
db = myclient["local"]
for name,url in files:
    db.drop_collection(name)
    json_file = requireAPIs(url)
    if(json_file): db[name].insert_many(json_file)
myclient.close()        
```

### Load Data from File System and Push (Some of it) to MySQL

```python
#connect to mysql
host_name = "localhost"
host_ip = "127.0.0.1"
port = "3306"
user_id = "root_user"
pwd = "Eutocomfome,meuamigo!"
mysql_db = "DadosAbertos_sql"
def createDatabase():
    conn_str = f"mysql+pymysql://{user_id}:{pwd}@{host_name}/sys" #use default sys db to create db I'll use
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    with sqlEngine.connect() as conn:
        conn.execute(text(f"DROP DATABASE IF EXISTS `{mysql_db}`;"))
        conn.execute(text(f"CREATE DATABASE `{mysql_db}`;"))
        conn.execute(text(f"USE {mysql_db};"))
    
def set_dataframe2(df,db,table_name, db_operation,pk_column):
    conn_str = f"mysql+pymysql://{user_id}:{pwd}@{host_name}/{mysql_db}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    with sqlEngine.connect() as conn:
        if db_operation == "insert": 
            df.to_sql(table_name, con=conn, index=False, if_exists='replace')
            conn.execute(text(f"ALTER TABLE {table_name} ADD PRIMARY KEY ({pk_column})"))
        elif db_operation == "update": 
            df.to_sql(table_name, con=conn, index=False, if_exists='append')
def replaceURIwithID(df):
    return df['uri'].apply(lambda x: int(re.findall(r'[^/]+$', str(x))[0]))
```

### Fetch MongoDB

```python
def get_mongo_dataframe(connect_str, db_name, collection, query):
    '''Create a connection to MongoDB'''
    client = pymongo.MongoClient(connect_str)
    
    '''Query MongoDB, and fill a python list with documents to create a DataFrame'''
    db = client[db_name]
    dframe = pd.DataFrame(list(db[collection].find(query)))
    dframe.drop(['_id'], axis=1, inplace=True)
    client.close()
    return dframe
```

```python
eventos_df = get_mongo_dataframe(conn_str["local"],"local","eventos",{})
eventos_df.head(1)
```
